```python
import pandas as pd
```

# Read Quantitative Data

```python
df = pd.read_csv('cleaned_data/quant_features.csv')
df.head()
```

```python
df.shape
```

# Read Qualitative Data

```python
df2 = pd.read_csv('cleaned_data/airbnb.csv')
df2.head()
```

```python
df2.neighborhood_id = df2.neighborhood_id.str.upper()
df2.head()
```

```python
df2.shape
```

```python
df3 = pd.read_csv('cleaned_data/foursquare.csv')
df3.head()
```

# Merge Datasets on Unique Identifier

```python
merged = pd.merge(df,df2,left_on='community', right_on = 'neighborhood_id')
merged = pd.merge(merged,df3,left_on='neighborhood_id', right_on = 'Neighborhood')
merged.head()
```

```python
merged.shape
```

# Drop Duplicate Column and Split Qualitative/Quantitative Features

```python
print(merged.columns.tolist())
```

```python
merged.drop(['Neighborhood','community_x'],axis=1,inplace=True)
merged.shape
```

```python
# move example neighborhood/neighborhood to beginning
col = merged.pop("neighborhood_id")
merged.insert(0, col.name, col)
```
