```python
import pandas as pd
```

# Read Qualitative Data

```python
df = pd.read_csv('cleaned_data/airbnb.csv')
df.head()
```

```python
df.shape
```

# Read Quantitative Data

```python
df2 = pd.read_csv('cleaned_data/quant_features.csv')
df2.head()
```

```python
df2.shape
```

# Merge Datasets on Unique Identifier

```python
merged = pd.merge(df,df2,left_on='neighborhood_id', right_on = 'Unique Identifier')
merged.head()
```

```python
merged.shape
```

# Drop Duplicate Column and Split Qualitative/Quantitative Features

```python
print(merged.columns.tolist())
```

```python
merged.drop('Unique Identifier',axis=1,inplace=True)
merged.shape
```

```python
# move example neighborhood to beginning (with neighborhood_id)
col = merged.pop("neighborhood")
merged.insert(1, col.name, col)
```

```python
merged.columns[215]
```

```python
merged.columns[216]
```

- merged
- merged[2:215] are qualitative features
- merged[216:] are quantitative features
- merged [215] is example neighborhood for each unique ID

# Write Output File

```python
merged.to_csv('cleaned_data/merged.csv',index=False)
```
